```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

#    Smile Detection

```python
smile_face=cv2.CascadeClassifier(cv2.data.haarcascades+"haarcascades/haarcascades_smile.xml")
#or
smile_face = cv2.CascadeClassifier('data/haarcascades/haarcascade_smile.xml')
```

```python
img=cv2.imread('data/happy2.jpeg')
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
smile=smile_face.detectMultiScale(gray,3.0,18)
for (x, y, w, h) in smile:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 5)

plt.figure(figsize=(7,7))
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```
